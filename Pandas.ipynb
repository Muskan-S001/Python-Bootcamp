```python
class Human:
    def __init__(self, fullName, age, color):
        self.name = fullName
        self.age = age
        self.color = color
        
    def tellYourName(self):
        print("Hi my name is " + self.name)
    def tellYourAge(self):
        print("hi my age is " + str(self.age))
        
```

```python
h1 = Human("Muskan")
h2 = Human("Somendra")
```

```python
h1.tellYourName()
```

```python
h2.tellYourName()
```

```python
h3 = Human("Somadri", 28, "xyz")
```

```python
h3.tellYourAge()
```
